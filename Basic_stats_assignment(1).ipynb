## 1. Descriptive Analytics for Numerical Columns

#### Objective: To compute and analyze basic statistical measures for numerical columns in the dataset

```python
import pandas as pd
import numpy as np
df=pd.read_csv("sales_data_with_discounts.csv")
```

#### Numerical & Categorical columns

```python
numerical_colms=df.select_dtypes(include='number').columns.tolist()
categorical_colms=df.select_dtypes(include='object').columns.tolist()
print("Numerical Columns",numerical_colms)
print("Categorical columns",categorical_colms)
```

#### Descriptive Analytics for Numerical Columns

```python
for col in numerical_colms:
    print("\nStatistics of",col)
    print("Mean   :",df[col].mean())
    print("Median :",df[col].median())
    print("Mode   :",df[col].mode()[0])
    print("Std_dev:",df[col].std())
```

```python
### Interpretation of these statistics
# Volume: The Average quantity sold is about 5 units, with moderate variation in quantity sales.
# Average price: Average price is around 10,453, but the prices may vary widely.
# Total sales value: Very high variability, with an average sales around 33,800.
# Discount Rate: An average discount about 15% is given and it is fairly consistent with little variation.
# Discount amount: Average discount amount is 3,346, showing a large spread in discount amount.
# Net sales value: Average net revenue is about 30,466, but it varies a lot.
```

## 	2. Data Visualization

#### Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.

#### Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
for col in numerical_colms:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col],kde=True)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

#### Interpretation for Histogram

```python
## Volume histogram: 
# The distribution is right-skewed with a long tail towards higher volumes. 
# Most sales happen at lower volumes, but some orders have very large quantities, indicating occasional bulk purchases. 
# A few outliers are present.

## Avg Price histogram: 
# Highly right-skewed with a long tail for high prices. 
# Majority of prices cluster at lower values, but some products are priced significantly higher. 
# This suggests a diverse pricing strategy across products.

## Total Sales Value histogram: 
# Also right-skewed showing a concentration of sales around lower values with some very large transactions causing a long tail. 
# The tail reflects a few high-value sales orders.

## Discount Rate histogram: 
# The distribution looks more symmetric with a peak near moderate discount values, 
# showing most discounts hover around a typical percentage. 
# There are fewer extreme discount values, so outlier presence is limited.

## Discount Amount histogram: 
# Right-skewed with most discounts being small, 
# but some of high discount amounts also exist, acting as outliers.

## Net Sales Value histogram: 
# Again, the distribution is right-skewed, reflecting many moderate net sales transactions, 
# but with some large net sales orders appearing as outliers.
```

## 3. Standardization of Numerical Variables

#### Objective: To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
standardization=scaler.fit_transform(df[numerical_colms])
standardized_data=pd.DataFrame(standardization,columns=numerical_colms)
standardized_data.head()
```

```python
## Standardization:
# Standardization is also known as z-score normalization, 
# transforms data so that it has a mean of 0 and a standard deviation of 1.
```

```python
for col in numerical_colms:
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(df[col],kde=True)
    plt.title(f"Original {col} (Before)")

    plt.subplot(1,2,2)
    sns.histplot(standardized_data[col],kde=True,color='b')
    plt.title(f"\nStandardized {col} (After)")
    plt.tight_layout()
    plt.show()
```

#### Before and After standardizaiton comparision

```python
## Before standardization:
# Numerical variables have varied scales and wide ranges.
# Histograms display skewed and spread out distributions, 
# making direct comparison and machine learning application difficult.

## After standardization
# Distributions shift to have approximately zero mean and unit standard deviation.
# The scales of all numerical features become comparable.
# Data values are centered around 0, with spread well normalized.
# This transformation reduces the impact of extreme values and makes features appropriate for algorithms sensitive to scale.
```

## 4. Conversion of Categorical Data into Dummy Variables

#### Objective: To transform categorical variables into a format that can be provided to ML algorithms.

```python
df_encoded = pd.get_dummies(df, columns=categorical_colms)
df_encoded.head()
```

```python
## One-hot encoding:
# One-hot encoding changes categories into numbers so models can understand them. 
# It makes a new column for each category, marking it with 1 if present, or 0 if not. 
# This avoids confusing the model by treating categories as numbers with order.
```
